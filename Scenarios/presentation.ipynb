```python
import os, sys

home = os.path.dirname(os.getcwd())
sys.path.insert(0, home + '\\Preprocess')
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'
```

```python
import warnings
warnings.filterwarnings("ignore")

import random
import classifier as cl
import keras
from PIL import Image
import numpy as np
```

```python
model = keras.models.load_model(os.path.join(home, "Attributes\\CNN_scenario7.h5"))
```

```python
imgs_dir = os.path.join(home, "Presentation Images")

test_imgs = []

for root, dirs, files in os.walk(imgs_dir):
    for f in files:
        img = Image.open(os.path.join(imgs_dir, f))
        img = img.resize((96, 96))
        img = np.array(img)
        img = img[:, :, :3]
        test_imgs.append(img)
        
random.shuffle(test_imgs)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

for i in range(len(test_imgs)):
    plt.subplot(2, 3, i + 1)
    plt.axis('off')
    plt.imshow(test_imgs[i], cmap='gray')
    
plt.show()
```

```python
for img in test_imgs:
    cl.test_img(img, model)
```
